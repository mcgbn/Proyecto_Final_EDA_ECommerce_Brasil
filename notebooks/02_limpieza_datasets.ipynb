## Cambio formato a fechas ##

```python
# Creamos funcion para aplicar el cambio de fecha a los distintos dataframes

def cambiar_a_fecha (df):
        columnas_fechas= df.filter(like='_at').columns.tolist() + df.filter(like='_timestamp').columns.tolist()+ df.filter(like='_date').columns.tolist()
        for columna in columnas_fechas:
            df[columna] =pd.to_datetime(df[columna], format='%Y-%m-%d %H:%M:%S',errors='coerce')
        return df
```

```python
# Creamos funcion para aplicar el cambio de fecha a los distintos dataframes (sin format)

def cambiar_a_fecha (df):
        columnas_fechas= df.filter(like='_at').columns.tolist() + df.filter(like='_timestamp').columns.tolist()+ df.filter(like='_date').columns.tolist()
        for columna in columnas_fechas:
            df[columna] =pd.to_datetime(df[columna],errors='coerce')
        return df
```

Aplicamos la funcion a los distintos dataframes

```python
order_items= cambiar_a_fecha(order_items)
order_items.info()
```

```python
orders= cambiar_a_fecha(orders)
orders.info()
```

```python
order_reviews= cambiar_a_fecha(order_reviews)
order_reviews.info()
```

## Cambio texto columnas a Title ##

```python
def formatear_texto_title (df, columnas):
    for columna in columnas:
        if columna in df.columns:
            df[columna]=df[columna].str.strip().str.title()
    return df
```

```python
#informamos  cuales son las que hay que modificar

texto_columnas_title= ['customer_city','geolocation_city', 'seller_city', 'product_category_name', 'product_category_name_english', 'payment_type']
```

```python
customers=formatear_texto_title(customers, texto_columnas_title)
geolocation=formatear_texto_title(geolocation, texto_columnas_title)
products=formatear_texto_title(products, texto_columnas_title)
order_payments=formatear_texto_title(order_payments,texto_columnas_title)
sellers= formatear_texto_title(sellers, texto_columnas_title)
product_category_translation=formatear_texto_title(product_category_translation, texto_columnas_title)
```

### Quitamos barras bajas y sustituimos como espacios, eliminamos posibles espacios extra

```python
def quitar_barras_bajas_y_espacios_extra(df, columnas):
    for columna in columnas:
        if columna in df.columns:
            df[columna]=df[columna].str.replace('_',' ',regex=False).str.strip()
    return df
```

```python
texto_columnas_espacios= ['customer_state', 'seller_state', 'payment_type', 'product_category_name', 'product_category_name_english']
```

```python
customers=quitar_barras_bajas_y_espacios_extra(customers, texto_columnas_espacios)
geolocation= quitar_barras_bajas_y_espacios_extra(geolocation, texto_columnas_espacios)
order_payments= quitar_barras_bajas_y_espacios_extra(order_payments, texto_columnas_espacios)
sellers= quitar_barras_bajas_y_espacios_extra(sellers, texto_columnas_espacios)
products=quitar_barras_bajas_y_espacios_extra(products,texto_columnas_espacios)
product_category_translation=quitar_barras_bajas_y_espacios_extra(product_category_translation, texto_columnas_espacios)
```
